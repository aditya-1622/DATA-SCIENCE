```python
# Dot product :
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python
# dot product : (1*4) + (2*5) + (3*6) = 4 + 10 + 18 = 32
```

```python
print(np.dot(a,b))
```

```python
print(a @ b)
```

```python
a2 = np.array([[1, 2],
     [3, 4]])

b2 = np.array([[5, 6],
     [7, 8]])
```

```python
# First row of a2 x first column of b2
# (1*5) + (2*7) = 19
# First row of a2 x second col of b2
# (1*6) + (2*8) = 22
```

```python
# Second row of a2 x first column of b2
# (3*5) + (4*7) = 43
# Second row of a2 x second column of b2
# (3*6) + (4*8) = 50
```

```python
print(a2 @ b2)
```
