```python
# EDA : Exploratory Data Analysis!
# What is EDA : which involves summarizing data & vizualising data to understand its main characterstics.
# Helps in identifying patterns, trends, relationships and anomalies in the data.
#  This makes data interpretation easier and helps in better decision-making.
```

```python
# Key objectives of EDA :
# 1) Understanding data distribution
# 2) Identifying missing values & anomalies
# 3) Viz. relationships
# 4) Detecting patterns
# 5) Helps in better decision-making
```

```python
# Types of Analysis in EDA :
# 1) Uni   Variate Analysis : Analysing single var (Histogram, boxplots, piecharts, etc.)
# 2) Bi    Var iate Analysis : Analysing 2 vars. (Scatterplots, Line Charts, boxplots, bar charts,etc.)
# 3) Multi Variate Analysis : Analysing 2+ vars togther (Pair Plots, heatmaps, 3d scatter plots, etc.)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "age": [22,25,28,35,40,45,50,29,31,38],
    "salary": [25000,30000,35000,50000,65000,70000,85000,40000,45000,60000],
    "experience": [1,2,3,7,10,12,15,4,5,8],
    "department": ["HR","IT","IT","Finance","Finance","IT","HR","IT","Finance","HR"]
})
```

```python
df
```

```python
# matplotlib --> syntax :
# plt.figure(figsize = (10,10)) # Creates chart area with size!
# plt.plot(x,y) # Draw Chart --> IMPORTANT!
# plt.title() # Add chart title
# plt.xlabel() # Add label to x-axis
# plt.ylabel() # Add label to y-axis
# plt.show() # Display
```

```python
# 1) Histogram : Show distribution of a numerical variable.
```

```python
plt.figure(figsize = (5,5)) # Creates chart area with size!
plt.hist(df["salary"], bins = 5)  # it becomes histogram
plt.title("Distribution of Salary") # add chart to title
plt.xlabel("Salary") # add label to x axis
plt.ylabel("Frequency") # add label to y axis
plt.show()  # display
```

```python
# 2) BoxPlot -  distribution of data, spread, outliers
plt.boxplot(df['salary'])
plt.show()
```

```python
# 3) Line Chart : are used to viz trends over time.
# X-axis values
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
# Y-axis values
monthly_sales = [100, 120, 140, 160, 180, 200]
# Plot line chart
plt.plot(months, monthly_sales)
# Display chart
plt.show()

#this is example of bi variate analysis.
```

```python
# 4) Scatter plot : analyse relationship between two numerical variables!
```

```python
df.head()
```

```python
plt.scatter(df['experience'],df['salary'])
plt.title("Experience vs Salary")
plt.xlabel("Experience")
plt.ylabel("Salary")

plt.show()
```

```python
plt.scatter(df['age'],df['salary'])
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")

plt.show()
```

```python
# 5) Bar Charts : Comparing values across categories!
```

```python
df
```

```python
avg_salary = df.groupby("department")["salary"].mean().reset_index()
avg_salary
```

```python
plt.bar(avg_salary["department"], avg_salary["salary"])
plt.title("Average salary by department")
plt.xlabel("Department")
plt.ylabel("Average salary")
plt.show()
```

```python
# 6) PieChart
dept_counts = df['department'].value_counts()
dept_counts
```

```python
dept_counts.values
```

```python
dept_counts.index
```

```python
plt.pie(dept_counts, labels = dept_counts.index , autopct = "%1.1f%%")
plt.show()
```

```python
# Correlation : Strength of linear assoc. between 2 numerical variables.
# Range --> [-1,+1] --> [high neg corr, high pos corr]
```

```python
corr = df[['age','salary','experience']].corr()
corr

# but it is difficult to read and for that we use

# heatmap --> seaborn!
```

```python
# 8) Bubble Chart :
# x -> exp
# y -> salary
# bubble size -> age
# bubble color -> department

#it is like scatter but varaiables use more here.
#this is multi variate
```

```python
df = pd.DataFrame({
    "age": [22,30,40,50,60,90,50,29,31,38],
    "salary": [25000,30000,35000,50000,65000,70000,85000,40000,45000,60000],
    "experience": [1,2,3,7,10,12,15,4,5,8],
    "department": ["HR","IT","IT","Finance","Finance","IT","HR","IT","Finance","HR"]
})
```

```python
plt.scatter(df["experience"], df["salary"], s = df["age"],
        c = df["department"].map({"HR" : "Blue", "IT": "Red", "Finance": "Green"}))
plt.show()
```
