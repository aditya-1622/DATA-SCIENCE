```python
# Regular Expression (re) : way to define pattern in the text.
# re module in python helps us finding, matching, manipulaitng text!
```

```python
# Phone number :
# +91 9999 123456
# +91-9999-123-456
# +919999123456
# +919999-123-456

# the only correct format is +91-4digits-3digits-3digits
```

```python
import re
```

```python
# 1. re.search()   -> search for the first match of pattern in a string
# 2. re.findall()  -> returns all non overlapping matches as a list
# 3. re.match()    -> match the pattern only at the start of the string
# 4. re.sub()      -> replaces part of the string
# 5. re.compile()  -> reusable object
```

```python
# Example 1 : Find all word-like sequences (alphanumeric + underscore)
text = "Hello_World123, how are you?"
```

```python
re.findall(r'\w+',text)
```

```python
# 1. Find all words in the sentence
# Extract all words (alphanumeric + underscore) from this string:
text = "Welcome_to Regex101! Learn & practice now."
re.findall(r'\w+',text)
```

```python
# 2. Check if a string starts with 'Hi'
text = "Hi there!"
re.findall(r'^Hi',text)
```

```python
# 3. Find all digits in the string
text = "Room number 42, Floor 5"
re.findall(r'\d+',text)
```

```python
# 4. Match if string ends with a question mark
text = 'Are you coming?'
re.findall(r'\?$',text)
```

```python
# 5.
text = "My number is 9876543210 and not 98765432101"
# 9876543210 : 10 digit number
# 98765432101 : 11 digit number
# Can you extract the correct mobile number from this?
re.search(r'\d{10}',text).group()
```
