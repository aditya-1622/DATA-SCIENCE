```python
df = pd.DataFrame({
    "age": [25, 30, np.nan, 22, np.nan],
    "salary": [50000, np.nan, 60000, np.nan, 45000],
    "city": ["BLR", "DEL", np.nan, "MUM", "BLR"],
    "gender": ["M", np.nan, "F", "F", np.nan],
    "purchased": [1, 0, 1, 0, 1]
})

print(df)
```

```python
# Loop through each column and:
# If numeric → fill missing (np.nan) with mean
# If categorical → fill missing with "Unknown"
```

```python
# Loop through each column
for col in df.columns:

    print(col, "-", df[col].dtype)

    # If numeric column
    if df[col].dtype != "object":

        # Fill missing values with mean
        mean_value = df[col].mean()
        df[col] = df[col].fillna(mean_value)

    else:
        # If categorical column
        # Fill missing values with "Unknown"
        df[col] = df[col].fillna("Unknown")
```

```python
df = pd.DataFrame({
    "age": [25, np.nan, 40, 22, np.nan],
    "income": [50000, 60000, np.nan, 45000, np.nan],
    "city": ["BLR", np.nan, "MUM", "DEL", np.nan],
    "gender": ["M", "F", np.nan, "F", np.nan],
    "is_member": [1, 0, 1, 1, 0]   # binary column
})

print(df)
```

```python
# Q. Conditional Filling with Special Rules

# Loop through each column and apply the following logic:
# If column is numeric:
  # If it is a binary column (only 0 and 1) → DO NOT change it
  # Otherwise → fill missing values with median
# If column is categorical (object):
  # Fill missing values with the most frequent value (mode)
# Return the cleaned DataFrame
```

```python
for col in df.columns:

    # Check if column is numeric
    if df[col].dtype != "object":

        # Get unique non-null values
        unique_vals = df[col].dropna().unique()

        # Check if binary column (only 0 and 1)
        if set(unique_vals).issubset({0, 1}):
            continue

        else:
            # Fill numeric missing values with median
            median_value = df[col].median()
            df[col] = df[col].fillna(median_value)

    else:
        # Fill categorical missing values with mode
        mode_value = df[col].mode()[0]
        df[col] = df[col].fillna(mode_value)

print(df)
```

```python
df = pd.DataFrame({
    "customer": ["A","B","C","D","E","F"],
    "amount": [100, 120, 130, 110, 115, 50000]  # 5000 is outlier
})

print(df)
```

```python
df["amount"].mean()
```

```python
df["amount"].median()
```

```python
#DETECT OUTLIERS.

# OPTION 1: BUSINESS RULE
#AMOUNT CANNOT BE GREATER THAN 1000
```

```python
# Option 2 : Using Stats! (Capping)
# Lower = Q1 - 1.5 * IQR
# Upper = Q3 + 1.5 * IQR
# IQR = Q3 - Q1 (75%ile - 25%ile)

#most used option

df['amount']
```

```python
df["amount"].median()
```

```python
Q1 = df['amount'].quantile(0.25)
Q3 = df['amount'].quantile(0.75)
IQR = Q3-Q1
```

```python
print(Q1)
print(Q3)
print(IQR)
```

```python
lower = Q1 - 1.5 * IQR #BOUND
upper = Q3 + 1.5 * IQR
```

```python
print(lower)
print(upper)
```

```python
# 2a. Filter
df_clean = df[(df["amount"] >= lower) & (df["amount"] <= upper)]
```

```python
# 2b. Capping
# Values < lower → set to lower
# Values > upper → set to upper
df['amount'] = np.clip(df['amount'],lower,upper)
```
