```python
# Q12. Time-based Aggregation (Important Real Use Case)

# The business wants to track monthly revenue trends.

# Steps:

# Extract month_name from order_date
# Group the data by month_name
# Calculate total amount per month
# Sort months in calendar order (NOT alphabetical)

# Final output should clearly show:

# month_name | total_revenue
```

```python
# ============================================================
# Q1. Date Handling
# Convert string → datetime and extract useful features
# ============================================================

# Convert to datetime format (VERY important before using .dt)
df["order_date"] = pd.to_datetime(df["order_date"])

# Extract components
df["year"] = df["order_date"].dt.year
df["month"] = df["order_date"].dt.month
df["day"] = df["order_date"].dt.day

# Extract readable month name (January, February…)
df["month_name"] = df["order_date"].dt.month_name()

# Check if weekend (Saturday=5, Sunday=6)
df["is_weekend"] = df["order_date"].dt.dayofweek >= 5


# ============================================================
# Q2. Apply Logic
# Apply custom business rule on each row
# ============================================================
# Define a function for business logic
def categorize_spend(x):
    if x >= 400:
        return "High"
    elif x >= 200:
        return "Medium"
    else:
        return "Low"

# Apply the function to the column
df["spend_type"] = df["amount"].apply(categorize_spend)

# ============================================================
# Q3. Map + Replace
# Transform categorical values
# ============================================================

# Map city names → codes (1-to-1 mapping)
city_map = {
    "Bangalore": "BLR",
    "Delhi": "DEL",
    "Mumbai": "MUM",
    "Chennai": "CHE"
}
df["city_code"] = df["city"].map(city_map)

# Replace values inside column (cleaning / encoding)
df["category"] = df["category"].replace({
    "Electronics": "E",
    "Clothing": "C"
})


# ============================================================
# Q4. Pivot (Aggregation + Reshape)
# Customer vs Category → Total spend
# ============================================================

pivot_df = df.pivot_table(
    index="customer",        # rows
    columns="category",      # columns
    values="amount",         # values to aggregate
    aggfunc="sum"            # aggregation function
)


# ============================================================
# Q5. Melt (Wide → Long)
# Convert pivot back to long format
# ============================================================

melt_df = pd.melt(
    pivot_df.reset_index(),  # bring index back as column
    id_vars=["customer"],    # columns to keep fixed
    var_name="category",     # new column name (from headers)
    value_name="amount"      # new column for values
)


# ============================================================
# Q6. Vectorization (Efficient computation)
# Apply discount without using apply()
# ============================================================

# First apply 10% discount to all
df["discounted_amount"] = df["amount"] * 0.9

# Then overwrite where condition is not met
df.loc[df["amount"] <= 300, "discounted_amount"] = df["amount"]


# ============================================================
# Q7. Pipeline (Date → Pivot → Melt)
# End-to-end transformation
# ============================================================

# Extract month name
df["month"] = df["order_date"].dt.month_name()

# Pivot: month vs category → total amount
temp = df.pivot_table(
    index="month",
    columns="category",
    values="amount",
    aggfunc="sum"
)

# Melt back to long format
q7 = pd.melt(
    temp.reset_index(),
    id_vars=["month"],
    var_name="category",
    value_name="total_amount"
)


# ============================================================
# Q8. BONUS (City × Month Revenue)
# Group and aggregate
# ============================================================

# Group by city and month → total revenue
q8 = df.groupby(["city", "month"])["amount"].sum().reset_index()

# Sort highest revenue first
q8 = q8.sort_values(by="amount", ascending=False)


# ============================================================
# Q9. Pivot with Multiple Aggregations
# Get both sum and mean
# ============================================================

q9 = df.pivot_table(
    index="customer",
    columns="category",
    values="amount",
    aggfunc=["sum", "mean"]  # multiple aggregations
)


# ============================================================
# Q10. Filter + Transformation
# High-value transactions
# ============================================================

# Filter rows
q10 = df[df["amount"] >= 300].copy()

# Add flag column
q10["high_value_flag"] = "Yes"

# Select required columns
q10 = q10[["customer", "city", "amount", "high_value_flag"]]


# ============================================================
# Q11. Conditional Logic (Vectorized, no apply)
# Categorize order size
# ============================================================

conditions = [
    df["amount"] >= 400,
    (df["amount"] >= 200) & (df["amount"] < 400),
    df["amount"] < 200
]

choices = ["Large", "Medium", "Small"]

# Apply vectorized conditional logic
df["order_size"] = np.select(conditions, choices)


# ============================================================
# Q12. Monthly Revenue (Correct Calendar Order)
# ============================================================

# Group by month
q12 = df.groupby("month_name")["amount"].sum().reset_index()

# Define correct calendar order (important!)
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", "December"]

# Convert to ordered categorical for sorting
q12["month_name"] = pd.Categorical(q12["month_name"], categories=month_order, ordered=True)

# Sort properly (not alphabetical)
q12 = q12.sort_values("month_name")
```
