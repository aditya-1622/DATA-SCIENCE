```python
#COMPARES LISTS VS ARRAY: PERFORMANCE..

import time
lst = list(range(100000))
start = time.time()
lst = [x * 2 for x in lst]
end = time.time()
print("List Time:", end - start)
```

```python
import time
arr = np.arange(100000)
start = time.time()
arr = arr * 2
end = time.time()
print("Numpy Time:", end - start)
```

```python
lst = [10,0,6,11,19,0,100,0,0,1]

#shift all zerores towards the end.
# order of non zero elements should not change

zero = []
nonzero = []
for i in lst:
  if i == 0:
    zero.append(i)
  else:
    nonzero.append(i)
print(nonzero + zero)
```

```python
lst = [10,0,6,11,19,0,100,0,0,1]
non_zeros = []
zeros_count = 0
for element in lst:
  if element != 0:
   non_zeros.append(element)
  else:
    zeros_count +=1

  lst = non_zeros + [0] * zeros_count
  print(lst)
```
