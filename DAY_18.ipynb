```python
import pandas as pd

# Sample Dataset
data = {
    "name": ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"],
    "department": ["IT", "HR", "IT", "HR", "Finance", "IT", "Finance", "HR", "IT", "Finance"],
    "city": ["BLR", "DEL", "BLR", "DEL", "BLR", "DEL", "BLR", "DEL", "BLR", "DEL"],
    "salary": [60000, 50000, 70000, 45000, 80000, 65000, 75000, 48000, 72000, 82000],
    "experience": [2, 3, 5, 2, 7, 4, 6, 3, 5, 8]
}

df = pd.DataFrame(data)
df
```

```python
# Q1: Find total salary paid per department
df.groupby('department')['salary'].sum()
```

```python
# Q2: Find average experience per city

df.groupby('city')['experience'].mean()
```

```python
# Q3: Find minimum salary in each department

df.groupby('department')['salary'].min()
```

```python
# Q4: Group by department and city → find total salary
df.groupby(['department','city'])['salary'].sum()
```

```python
# Q5: Group by department → find mean salary AND max experience together
df.groupby('department').agg(
    {
        'salary' : 'mean',
        'experience' : 'max'
    }
)
```

```python
# Pivot Table : & Group by both are same you can use as per your easy.
# pd.pivot_table(df, values=..., index=..., columns=..., aggfunc=...)
# df -> Dataset
# values -> column you want to aggregate
# index -> "department" (grouping level 1)
# column -> creates a matrx (grouping level 2)
# aggfunc -> what you want to do with the values (sum/mean/min/max/median)
```

```python
df.head()
```

```python
# Avg salary for each dept!
pd.pivot_table(df, values = 'salary', index='department', aggfunc='mean')
```

```python
# Dept vs City salary comparison
pd.pivot_table(df, values = 'salary', index='department', columns='city' ,
               aggfunc='mean')
```

```python
df.groupby(['department','city'])['salary'].mean()
# similar as above
```
