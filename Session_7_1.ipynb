```python
# Customers DataFrame
df_customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4],
    "order_id": ["A", "B", "C", "D"],
    "city": ["Bangalore", "Delhi", "Mumbai", "Chennai"]
})

# Orders DataFrame
df_orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104],
    "customer_id": [1, 2, 3, 5],
    "amount": [500, 300, 200, 700]
})

print(df_customers)
print('-'*32)
print(df_orders)
```

```python
pd.merge(df_customers,df_orders, how='inner',on='customer_id')
```

```python
pd.merge(df_customers,df_orders, how='inner',on='customer_id',
         suffixes=('_left','_right'))
```

```python
df1=pd.DataFrame(
    {"letter":["a","b","c"],
     "num":range(3)})
df2=pd.DataFrame(
    {"value":[3,5,7]},
    index=["a","b","e"])
```

```python
df1 = pd.DataFrame({
    "name": ["A", "B", "C"],
    "score": [80, 85, 90]
})

df2 = pd.DataFrame({
    "name": ["D", "E"],
    "score": [70, 75]
})

print(df1)
print('-'*32)
print(df2)
```

```python
# Row-wise
pd.concat([df1,df2], axis = 0)
```

```python
pd.concat([df1,df2], axis = 0,ignore_index=True)
```

```python
pd.concat([df1,df2], axis = 1)
# can add - ignore_index=True
```
