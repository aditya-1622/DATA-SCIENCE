```python
## Regular Expression (re) : way to define pattern in the text.
# re module in python helps us finding, matching, manipulaitng text!
```

```python
# Phone number :
# +91 9999 123456
# +91-9999-123-456 @this is correct way
# +919999123456
# +919999-123-456

# the only correct format is +91-4digits-3digits-3digits

```

```python
#now how yu write this command and run it
import re
```

```python
# 1. re.search()   -> search for the first match of pattern in a string
# 2. re.findall()  -> returns all non overlapping matches as a list
# 3. re.match()    -> match the pattern only at the start of the string
# 4. re.sub()      -> replaces part of the string
# 5. re.compile()  -> reusable object
```

```python
# Example 1 : Find all word-like sequences (alphanumeric + underscore)
text = "Hello_World123, how are you?"
```

```python
re.findall(r"\w", text)
# \w  : word chars (letter, digits, underscore)
# and removes question marks and spaces
```

```python
re.findall(r"\w+",text)
#w+: means one or repitions of \w
```

```python
text = 'Masai School is 10 year old!'
# I want to find position of only digits in the text?
```

```python
re.search(r"\d",text)

# \d tells us on which place from counting left our digits lies we can simply find it by reading but wehn we works on
# a large data set it is dificult to find and here \d comes into picture.
```

```python
re.search(r"\d+",text)
```

```python
re.findall(r'\d+',text)
#now we don't want a position we want entire digit and for that we use (\d+)
```
