```python
# Why use seaborn?
# 1) Less code
# 2) Default styling
# 3) Works directly with dfs
# 4) Easy to create viz.
```

```python
import numpy  as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# matplotlib :
# plt.scatter(df['x'],df['y'])

# seaborn :
# sns.scatterplot(x = "", y ="", df = "") # cleaner and more readable
```

```python
df = pd.DataFrame({
    "age": [22,25,28,35,40,45,50,29,31,38],
    "salary": [25000,30000,35000,50000,65000,70000,85000,40000,45000,60000],
    "experience": [1,2,3,7,10,12,15,4,5,8],
    "department": ["HR","IT","IT","Finance","Finance","IT","HR","IT","Finance","HR"]
})

df
```

```python
# sns.chart_type(x='',y='',data='')
```

```python
# 1) Histogram : Distribution of numerical variables.
sns.histplot(data=df,x='salary',bins=7)
plt.title('Distribution of Salary')
plt.show()
```

```python
# 2) Boxplot : For distribution and outlier analysis
sns.boxplot(data=df,x='salary') # distribution of salary
sns.boxplot(data=df,x='salary',y='department')  # distribution of salary for each department
```

```python
# 3) Scatter plot : Relationship between 2 numerical variables.
sns.scatterplot(data = df ,x = 'salary', y = 'experience')
```

```python
# 4) Bubble chart : Scatter plot + third variable (multivariate)
sns.scatterplot(data = df ,x = 'salary', y = 'experience',
                hue='department', size = 'age')
# plt.legend(loc='upper right') # position of legends
```

```python
# 5) Line plot : Trends over time
sns.lineplot(data = df, x ='experience',y='salary')
```

```python
# 6) Bar Plot : Comparing aggregated category values!
sns.barplot(data=df,x='department',y='salary') # default estimator = mean
```

```python
sns.barplot(data=df,x='department',y='salary',
            estimator=np.max)
```

```python
# 7) : Count plot : counting category frequency
sns.countplot(data=df,x='department')
```

```python
# 8) Violin plot : Distribuion and shape!
sns.violinplot(data=df,x='department',y='salary')
```

```python
# 10) Heatmap : Correlation viz. (strength of linear assoc. between numerical values)
# Range of corr : -1 to +1 (high neg corr, high pos corr)
corr = df.corr(numeric_only=True)
corr
```

```python
# 12) SubPlots : helps you in comparing multiple charts side by side!
# fig,ax = plt.subplots(rows,cols)
# fig = entire figure
# ax  = individual subplot axes
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,4)) #1 row, 2 cols
ax[0].hist(df['salary'])
ax[1].boxplot(df['salary'])
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10,8))

# Top-left
ax[0,0].hist(df["salary"])
ax[0,0].set_title("Histogram")

# Top-right
ax[0,1].boxplot(df["salary"])
ax[0,1].set_title("Boxplot")

# Bottom-left
ax[1,0].scatter(df["experience"], df["salary"])
ax[1,0].set_title("Scatter")

# Bottom-right
ax[1,1].bar(
    df["department"],
    df["salary"]
)
ax[1,1].set_title("Bar Chart")

plt.tight_layout()

plt.show()
```

```python
np.random.seed(42)

n = 50

df = pd.DataFrame({
    "employee_id": range(1, n+1),

    "age": np.random.randint(22, 55, n),

    "experience": np.random.randint(1, 20, n),

    "salary": np.random.randint(25000, 120000, n),

    "department": np.random.choice(
        ["HR", "IT", "Finance", "Sales", "Marketing"],
        n
    ),

    "city": np.random.choice(
        ["Bangalore", "Delhi", "Mumbai", "Chennai"],
        n
    ),

    "performance_score": np.random.randint(50, 100, n),

    "work_hours": np.random.randint(5, 12, n),

    "remote_work": np.random.choice(
        ["Yes", "No"],
        n
    )
})
```

```python
# For each question:
# Decide the BEST chart to use
# Create the visualization
# Write 2–3 insights from the chart
```

```python
# Q1. Analyze the relationship between experience and salary.
# - Is there a positive or negative relationship?
# - Are there any outliers?

# Q2. Study the distribution of salary.
# - Is the data skewed?
# - Are most employees concentrated in a specific salary range?

# Q3. Compare the average salary across departments.
# - Which department earns the most?
# - Which department earns the least?

# Q4. Analyze the distribution of performance_score.
# - Are scores concentrated within a narrow range?
# - Are there any unusual observations?

# Q5. Study the relationship between work_hours and performance_score.
# - Does working more hours improve performance?

# Q6. Compare salary distributions across cities.
# - Which city has the widest salary spread?
# - Are there any salary outliers?

# Q7. Analyze department frequency.
# - Which department has the highest number of employees?

# Q8. Study the relationship between age and salary.
# - Do older employees generally earn more?

# Q9. Analyze the correlation between:
# - age
# - experience
# - salary
# - performance_score
# - work_hours

# - Which variables are strongly correlated?
# - Are there any negative correlations?

# Q10. Compare remote_work and performance_score.
# - Do remote employees perform differently from non-remote employees?
```

```python
# ============================================================
# Q1. Experience vs Salary
# Recommended Chart: Scatter Plot
# ============================================================

sns.scatterplot(
    x="experience",
    y="salary",
    data=df
)

plt.title("Experience vs Salary")
plt.show()



# ============================================================
# Q2. Salary Distribution
# Recommended Chart: Histogram
# ============================================================

sns.histplot(
    x="salary",
    data=df,
    bins=10
)

plt.title("Salary Distribution")
plt.show()



# ============================================================
# Q3. Average Salary Across Departments
# Recommended Chart: Bar Plot
# ============================================================

sns.barplot(
    x="department",
    y="salary",
    data=df
)

plt.title("Average Salary by Department")
plt.show()



# ============================================================
# Q4. Performance Score Distribution
# Recommended Chart: Histogram
# ============================================================

sns.histplot(
    x="performance_score",
    data=df,
    bins=10
)

plt.title("Performance Score Distribution")
plt.show()



# ============================================================
# Q5. Work Hours vs Performance Score
# Recommended Chart: Scatter Plot
# ============================================================

sns.scatterplot(
    x="work_hours",
    y="performance_score",
    data=df
)

plt.title("Work Hours vs Performance Score")
plt.show()



# ============================================================
# Q6. Salary Distribution Across Cities
# Recommended Chart: Boxplot
# ============================================================

sns.boxplot(
    x="city",
    y="salary",
    data=df
)

plt.title("Salary Distribution Across Cities")
plt.show()



# ============================================================
# Q7. Department Frequency
# Recommended Chart: Count Plot
# ============================================================

sns.countplot(
    x="department",
    data=df
)

plt.title("Department Frequency")
plt.show()



# ============================================================
# Q8. Age vs Salary
# Recommended Chart: Scatter Plot
# ============================================================

sns.scatterplot(
    x="age",
    y="salary",
    data=df
)

plt.title("Age vs Salary")
plt.show()



# ============================================================
# Q9. Correlation Between Numerical Variables
# Recommended Chart: Heatmap
# ============================================================

corr = df[[
    "age",
    "experience",
    "salary",
    "performance_score",
    "work_hours"
]].corr()

sns.heatmap(
    corr,
    annot=True
)

plt.title("Correlation Heatmap")
plt.show()



# ============================================================
# Q10. Remote Work vs Performance Score
# Recommended Chart: Boxplot
# ============================================================

sns.boxplot(
    x="remote_work",
    y="performance_score",
    data=df
)

plt.title("Remote Work vs Performance Score")
plt.show()
```
