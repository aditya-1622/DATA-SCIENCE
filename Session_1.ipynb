```python
# Importing important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the data
df = pd.read_csv("C:/Users/jaipu/Documents/Masai/New/Statistics_&_Data_Visualization/Content/Data/employee_salaries.csv")
df.head()
```

```python
# Getting an idea about the shape of the data
print('Shape :', df.shape)
print('Number of rows :', df.shape[0])
print('Number of columns :', df.shape[1])
```

```python
# Getting an idea about the proportion of employees in "department" column
df['department'].value_counts()
```

```python
# Getting the central tendency values for 'annual_salary_lpa'
print('Mean Salary :', df['annual_salary_lpa'].mean())
print('Median Salary :', df['annual_salary_lpa'].median())
print('Mode Salary :', df['annual_salary_lpa'].mode()[0])
```

```python
# As we can see, mean > median, indicating that the data is right skewed.
```

```python
# Plot
plt.figure(figsize = (5, 3))
plt.hist(df['annual_salary_lpa'], bins = 30)
plt.title('Distribution of Salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Measures of spread / dispersion
print('Range of Salary :', df['annual_salary_lpa'].max() - df['annual_salary_lpa'].min())
print('Variance of Salary :', df['annual_salary_lpa'].var())
print('Standard Deviation :', df['annual_salary_lpa'].std())
print('Inter Quartile Range (IQR) :', df['annual_salary_lpa'].quantile(0.75) - df['annual_salary_lpa'].quantile(0.25))
```

```python
# Statistical Summary
df.describe()
```

```python
# Shape -> Skewness, Kurtosis
print('Skewness :', df['annual_salary_lpa'].skew())
print('Kurtosis :', df['annual_salary_lpa'].kurt())
```

```python
# Boxplot
plt.figure(figsize = (5, 3))
plt.boxplot(df['annual_salary_lpa'], vert = False)
plt.title('Box Plot for Annual Salary')
plt.show()
```
