```python
import pandas as pd
import numpy as np

# Customers
customers = pd.DataFrame({
    "customer_id": [1, 2, 3, 4, 5],
    "name": ["A", "B", "C", "D", "E"],
    "city": ["Bangalore", "Delhi", "Mumbai", "Chennai", "Hyderabad"]
})

# Orders
orders = pd.DataFrame({
    "order_id": [101, 102, 103, 104, 105, 106],
    "customer_id": [1, 2, 2, 3, 6, 2],
    "amount": [500, 300, 200, 400, 800, 150]
})

# Payments
payments = pd.DataFrame({
    "payment_id": [201, 202, 203, 204],
    "order_id": [101, 102, 104, 107],
    "status": ["Paid", "Pending", "Paid", "Failed"]
})

```

```python
""" Q1. Inner Join Basics
Merge customers and orders to show only customers who have placed orders.
Output: customer_id, name, order_id, amount

Q2. Left Join (Unmatched Handling)
Show all customers and their orders (if any).
Identify customers who never placed an order

Q3. Right Join Insight
Show all orders with customer details.
Find orders that do not have a valid customer

Q4. Full Outer Join Analysis
Perform an outer join between customers and orders.
Count how many rows have missing values
Explain why they occur

Q5. Multi-table Join (Real-world scenario)
Combine customers, orders, and payments
Show: name, order_id, amount, payment status
Hint: You’ll need 2 merges

Q6. Cross Join Challenge 🔥
Create all possible combinations of:
customers × payment status
Questions:
Total number of rows?
Why is this useful?

Q7. Overlapping Columns + Cleanup
Create a new DataFrame:
customer_info = customers.rename(columns={"city": "location"})
Now:
Merge customers and customer_info
Handle duplicate columns properly
Clean final output (no _x, _y)
"""
```

ANSWERS..


```python
#Q1.
""" 1. Inner Join Basics
Merge customers and orders to show only customers who have placed orders.
Output: customer_id, name, order_id, amount
pd.merge(customers, orders, how = "inner" , on = "customer_id") """

pd.merge(customers, orders, how = "inner", on = "customer_id")
```

```python
 # Q2. Left Join (Unmatched Handling)
 # Show all customers and their orders (if any).
 # Identify customers who never placed an order




pd.merge(customers, orders, how = "left", on = "customer_id")
```

```python
 # Q3. Right Join Insight
 # Show all orders with customer details.
 # Find orders that do not have a valid customer

pd.merge(customers, orders, how = "right", on = "customer_id")
```

```python
#Q4. Full Outer Join Analysis
 # Perform an outer join between customers and orders.
 # Count how many rows have missing values
 # Explain why they occur

pd.merge(customers, orders, how = "outer", on = "customer_id")
```

```python
#Q5. Multi-table Join (Real-world scenario)
# Combine customers, orders, and payments
# Show: name, order_id, amount, payment status
# Hint: You’ll need 2 merges

# First merge: customers and orders
customers_orders = pd.merge(customers, orders, how = "left", on = "customer_id")

# Second merge: result with payments
final_df = pd.merge(customers_orders, payments, how = "left", on = "order_id")

# Select and display the required columns
print(final_df[['name', 'order_id', 'amount', 'status']])
```

```python
#Q6. Cross Join Challenge 🔥
# Create all possible combinations of:
# customers × payment status
# Questions:
# Total number of rows?
# Why is this useful?

# Extract unique payment statuses from the payments DataFrame
unique_payment_statuses = payments[['status']].drop_duplicates()

# Perform a cross join between customers and unique payment statuses
customer_payment_cross_join = pd.merge(customers, unique_payment_statuses, how='cross')

# Display the first few rows of the cross-joined DataFrame
display(customer_payment_cross_join.head())

# Calculate and display the total number of rows
total_rows = len(customer_payment_cross_join)
print(f"\nTotal number of rows: {total_rows}")

# Explain the usefulness of a cross join
```

```python
#Q7. Overlapping Columns + Cleanup
# Create a new DataFrame: customer_info
customer_info = customers.rename(columns={"city": "location"})

# Merge customers and customer_info
# We'll merge on customer_id, and handle the duplicated 'name' column.
merged_customers_info = pd.merge(customers, customer_info, on="customer_id", how="inner", suffixes=('', '_info'))

# The prompt asks for 'no _x, _y' but since we have two 'name' columns,
# pandas will add suffixes by default. Let's explicitly drop the duplicate 'name' from customer_info
# as it's identical to the one in 'customers', and keep 'city' and 'location' as distinct.

# First, let's merge with default suffixes to see what we get.
# Then, we'll select the columns we want.
merged_df_with_suffixes = pd.merge(customers, customer_info, on="customer_id", how="inner")

# To clean the output and avoid _x, _y for identical columns like 'name',
# we can decide which one to keep. Since both 'name' columns are identical,
# we'll keep the one from the 'customers' DataFrame (implicitly 'name_x').
# The 'city' and 'location' columns are semantically different due to renaming, so both are desired.
final_q7_df = merged_df_with_suffixes[['customer_id', 'name_x', 'city', 'location']]
final_q7_df = final_q7_df.rename(columns={'name_x': 'name'}) # Rename 'name_x' back to 'name'

print("Merged and cleaned DataFrame for Q7:")
display(final_q7_df)
```
