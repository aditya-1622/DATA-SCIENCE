```python
def validate_age(raw_age):
  # Rule 1 : it must exist.
  if raw_age is None :
    return False,"Age is missing"

  clean_age = int(raw_age)

# Rule 2 : logical bounds
  if clean_age < 0 or clean_age > 120 :
    return False, "Invalid age range"

# Rule 3 : clean data!
  return True, clean_age
```

```python
validate_age('28')
```

```python
validate_age(None)
```

```python
validate_age(-10)
```

```python
from pydantic import BaseModel, ValidationError, EmailStr, Field
```

```python
# Define the schema :
class UserSchema(BaseModel) :
  id   : int
  name : str = Field(min_length = 2)
  age  : int = Field(gt =0, lt = 120)
```

```python
# 1. good data
good_data = {
    'id'   : 123,
    'name' : 'masai',
    'age'  : 20,
}
```

```python
UserSchema(**good_data)
# dict unpacking!
```

```python
# 2. bad data
bad_data = {
    'id'   : 123,
    'name' : 'masai',
    'age'  : -20,
}
```

```python
UserSchema(**bad_data)
```

```python
# 1. Our simple function
def print_employee_badge(name, department, role):
    print(f"--- BADGE ---")
    print(f"Name: {name}")
    print(f"Dept: {department}")
    print(f"Role: {role}")
    print(f"-------------")
```

```python
employee_data = {
    "name": "Sarah Connor",
    "department": "Security",
    "role": "Director"
}
```

```python
print_employee_badge(
    name=employee_data["name"],
    department=employee_data["department"],
    role=employee_data["role"]
)
```

```python
print_employee_badge(**employee_data)
```
