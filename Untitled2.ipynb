```python
from datetime import datetime

class SignupSystem:
    def __init__(self):

        # Initialize the signup system. Use dictionary: key:email value: signup datetime

        self.signups = {}

    def add_signup(self, email: str) -> bool:

        #Add a new signup.
        #Returns True if added, False if duplicate.

        if email in self.signups:
            return False  # Duplicate

        self.signups[email] = datetime.now()
        return True

    def exists(self, email: str) -> bool:

      # Check if email already exists.
      # Time Complexity: O(1)

        return email in self.signups

    def get_first_100(self):

        # Retrieve first 100 signups in chronological order.
         # Time Complexity: O(100) ≈ O(1)

        return list(self.signups.items())[:100]

    def total_users(self) -> int:

     # Get total number of signups.

          return len(self.signups)

    def print_first_100(self):

    # Pretty print first 100 users.

        first_100 = self.get_first_100()
        for i, (email, date) in enumerate(first_100, start=1):
            print(f"{i}. {email} -> {date}")


# DEMO__TEST.
if __name__ == "__main__":
    system = SignupSystem()

    # Simulate adding users
    emails = [
        "sneha@gmail.com",
        "anshpandey@gmail.com",
        "adityapandey16@gmail.com",
        "anshpandey@gmail.com",  # duplicate
    ]

    for email in emails:
        if system.add_signup(email):
            print(f"✅ Added: {email}")
        else:
            print(f"❌ Duplicate: {email}")

    print("\n--- Existence Check ---")
    print(system.exists("anshpandey@gmail.com"))   # True
    print(system.exists("sneha@gmail.com")) # False

    print("\n--- First 100 Users ---")
    system.print_first_100()

    print("\n--- Total Users ---")
    print(system.total_users())
```
