```python
# Compares the performance of a list v/s array!
```

```python
import time
lst = list(range(1000000))
start = time.time()
lst = [x * 2 for x in lst]
end = time.time()
print("List time:", end - start)
```

```python
import time
arr = np.arange(1000000)
start = time.time()
arr = arr * 2
end = time.time()
print("NumPy time:", end - start)
```

```python
lst = [10,0,6,11,19,0,100,0,0,1]
# 1. shift all the 0s towards the end of the list
# 2. order of non zero elements should not change
# Output : [10,6,11,19,100,1,0,0,0,0]
```

```python
zero    = []
nonzero = []
for i in lst :
  if i == 0 :
    zero.append(i)
  else :
    nonzero.append(i)
```

```python
# Q. Write a program to print numbers from 1 to 20 with the following rules:
# If number is even → print "Even"
# If number is odd → print "Odd"
# If divisible by both 2 and 3 → print "EvenThree"
# Else print the number
```

```python
for i in range(1,21) :
  if i % 2 ==0 and i % 3 == 0 :
    print(i,"-","EvenThree")
  elif i % 2 == 0 :
    print(i,"-","Even")
  elif i % 2 != 0 :
    print(i,"-","Odd")
  else :
    print(i)
```
