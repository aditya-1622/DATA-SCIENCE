```python
# Q1: Given an integer x, return x with its digits reversed

def reverse_integer(x):
    # Check if number is negative
    sign = -1 if x < 0 else 1

    # Convert to string, reverse, convert back to int
    reversed_num = int(str(abs(x))[::-1])

    # Apply original sign
    return sign * reversed_num


# Examples
print(reverse_integer(123))   # Output: 321
print(reverse_integer(-123))  # Output: -321
```

```python
# Q2: Find 2nd largest number without using set

x = [1,2,3,100,100,100,99,98,97]

def second_largest(arr):
    largest = second = float('-inf')  # Initialize with very small numbers

    for num in arr:
        if num > largest:
            second = largest  # update second before changing largest
            largest = num
        elif num != largest and num > second:
            second = num  # update second only if not equal to largest

    return second

print(second_largest(x))  # Output: 99
```

```python
# Q3: Apply discount based on quantity

products = [(100, 5), (80, 10), (200, 15), (50, 7)]

def calculate_prices(products):
    result = []

    for price, quantity in products:
        total = price * quantity  # base price

        if quantity >= 10:
            total *= 0.8  # apply 20% discount

        result.append(int(total))  # convert to int for clean output

    return result

print(calculate_prices(products))  # Output: [500, 640, 2400, 350]
```

```python
# Q4: Reverse key-value pairs

dct = {1:"data", 2:"science"}

def reverse_dict(d):
    reversed_d = {}

    for key, value in d.items():
        reversed_d[value] = key  # swap key and value

    return reversed_d

print(reverse_dict(dct))  # Output: {'data': 1, 'science': 2}
```

```python
# Q5: Update dictionary using list of tuples

player_scores = {'Alice': 10, 'Bob': 8, 'Carol': 15}
updates = [('Alice', 5), ('Bob', 2), ('Dave', 7)]

def update_scores(scores, updates):
    for name, points in updates:
        if name in scores:
            scores[name] += points  # add to existing score
        else:
            scores[name] = points   # create new entry

    return scores

print(update_scores(player_scores, updates))
# Output: {'Alice': 15, 'Bob': 10, 'Carol': 15, 'Dave': 7}
```
