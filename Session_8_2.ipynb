```python
# EDA : Exploratory Data Analysis!
# What is EDA : which involves summarizing data & vizualising data to understand its main characterstics.
# Helps in identifying patterns, trends, relationships and anomalies in the data.
# This makes data interpretation easier and helps in better decision-making.
```

```python
# Key objectives of EDA :
# 1) Understanding data distribution
# 2) Identifying missing values & anomalies
# 3) Viz. relationships
# 4) Detecting patterns
# 5) Helps in better decision-making
```

```python
# Types of Analysis in EDA :
# 1) Uni   Variate Analysis : Analysing single var (Histogram, boxplots, piecharts, etc.)
# 2) Bi    Variate Analysis : Analysing 2 vars. (Scatterplots, Line Charts, boxplots, bar charts,etc.)
# 3) Multi Variate Analysis : Analysing 2+ vars togther (Pair Plots, heatmaps, 3d scatter plots, etc.)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "age": [22,25,28,35,40,45,50,29,31,38],
    "salary": [25000,30000,35000,50000,65000,70000,85000,40000,45000,60000],
    "experience": [1,2,3,7,10,12,15,4,5,8],
    "department": ["HR","IT","IT","Finance","Finance","IT","HR","IT","Finance","HR"]
})
```

```python
df
```

```python
# matplotlib --> syntax :
# plt.figure(figsize = (10,10)) # Creates chart area with size!
# plt.plot(x,y) # Draw Chart --> IMPORTANT!
# plt.title() # Add chart title
# plt.xlabel() # Add label to x-axis
# plt.ylabel() # Add label to y-axis
# plt.show() # Display
```

```python
# 1) Histogram : Show distribution of a numerical variable.
```

```python
plt.figure(figsize = (10,10)) # Creates chart area with size!
plt.hist(df['salary'],bins=5) # histogram
plt.title('Distribution of Salary') # Add chart title
plt.xlabel('Salary') # Add label to x-axis
plt.ylabel('Frequency') # Add label to y-axis
plt.show() # Display
```

```python
# 2) BoxPlot -  distribution of data, spread, outliers
plt.boxplot(df['salary'])
plt.show()
```

```python
# 3) Line Chart : are used to viz trends over time.
# X-axis values
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
# Y-axis values
monthly_sales = [100, 120, 140, 160, 180, 200]
# Plot line chart
plt.plot(months, monthly_sales)
# Display chart
plt.show()
```

```python
# 4) Scatter plot : analyse relationship between two numerical variables!
```

```python
df.head()
```

```python
plt.scatter(df['age'],df['salary'])
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")

plt.show()
```

```python
# 5) Bar Charts : Comparing values across categories!
```

```python
df
```

```python
avg_salary = df.groupby('department')['salary'].mean().reset_index()
avg_salary
```

```python
plt.bar(avg_salary['department'],avg_salary['salary'])
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.show()
```

```python
# 6) PieChart
dept_counts = df['department'].value_counts()
dept_counts
```

```python
dept_counts.values
```

```python
dept_counts.index
```

```python
plt.pie(dept_counts.values,labels = dept_counts.index, autopct = '%1.1f%%')
plt.show()
```

```python
# Correlation : Strength of linear assoc. between 2 numerical variables.
# Range --> [-1,+1] --> [high neg corr, high pos corr]
```

```python
corr = df[['age','salary','experience']].corr()
corr

# heatmap --> seaborn!
```

```python
# 8) Bubble Chart :
# x -> exp
# y -> salary
# bubble size -> age
# bubble color -> department
```

```python
df = pd.DataFrame({
    "age": [22,30,40,50,60,90,50,29,31,38],
    "salary": [25000,30000,35000,50000,65000,70000,85000,40000,45000,60000],
    "experience": [1,2,3,7,10,12,15,4,5,8],
    "department": ["HR","IT","IT","Finance","Finance","IT","HR","IT","Finance","HR"]
})
```

```python
plt.scatter(df['experience'],df['salary'],s = df['age'],
            c = df['department'].map({"HR" : "Blue", "IT" : "Red", "Finance":"Green"}))
plt.show()
```

| Chart Type | 1-Line Definition | Analysis Type | Variables Required |
|---|---|---|---|
| Histogram | Shows distribution of a numerical variable using frequency bins | Univariate | 1 Numerical |
| Boxplot | Shows spread, quartiles, and outliers of numerical data | Univariate / Bivariate | 1 Numerical (+ optional category) |
| Bar Chart | Compares values across categories | Univariate / Bivariate | 1 Categorical + 1 Numerical |
| Horizontal Bar Chart | Bar chart variant for better readability of long labels | Univariate / Bivariate | 1 Categorical + 1 Numerical |
| Pie Chart | Shows percentage contribution of categories to a whole | Univariate | 1 Categorical |
| Line Chart | Shows trends or changes over ordered/time data | Univariate / Bivariate | Ordered X + 1 Numerical |
| Scatter Plot | Shows relationship between two numerical variables | Bivariate | 2 Numerical |
| Bubble Chart | Scatter plot with bubble size representing a third variable | Multivariate | 3 Numerical (+ optional category/color) |
| Heatmap | Visualizes intensity/correlation using colors | Multivariate | Matrix / Multiple Numerical |
| Pairplot | Shows pairwise relationships among numerical variables | Multivariate | Multiple Numerical |
| Violin Plot | Shows both distribution shape and spread of numerical data | Univariate / Bivariate | 1 Numerical (+ optional category) |
| Area Chart | Emphasizes magnitude of trends over time | Bivariate | Ordered X + 1 Numerical |
| Stack Plot | Shows contribution of multiple groups over time | Multivariate | Ordered X + Multiple Numerical |
| Correlation Matrix | Displays correlation strength between variables | Multivariate | Multiple Numerical |
