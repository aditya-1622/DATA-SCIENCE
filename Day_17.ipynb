```python
df = pd.DataFrame({
    "Name": ["Amit", "Sara", "John", "Neha", "Raj", "Priya", "Arjun", "Meera", "David", "Kiran"],
    "ID": [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "City": ["Delhi", "Mumbai", "Pune", "Delhi", "Mumbai", "Bangalore", "Delhi", "Pune", "Mumbai", "Bangalore"],
    "Country": ["India", "India", "India", "India", "India", "India", "India", "India", "India", "India"],
    "Salary": [50000, 60000, 45000, 80000, 55000, 70000, 52000, 48000, 62000, 75000]
})

df
```

```python
# Q1 (AND condition)
# Get employees who are from Delhi AND have Salary > 50000
df[(df['City'] == 'Delhi') & (df['Salary']>50000)]
```

```python
# Q2 (OR condition)
# Get employees who are from Mumbai OR have Salary > 70000
df[(df['City'] == 'Mumbai') | (df['Salary']>70000)]
# | : pipe operation
```

```python
# Q3 (Greater than + Equal condition on different columns)
# Get employees where Salary >= 60000 AND City = Bangalore
df[(df.Salary >= 60000) & (df.City == "Bangalore")]
```

```python
# Q4 (NOT condition)
# Get employees who are NOT from Pune
df[df['City']!='Pune']
```

```python
# Q5 (Mixed condition – AND + OR)
# Get employees who:
# are from Delhi AND Salary > 50000
# OR
# are from Mumbai AND Salary >= 60000
# df[(() & ()) |
#    (() & ())]
df[
    ((df["City"] == "Delhi") & (df["Salary"] > 50000)) |
    ((df["City"] == "Mumbai") & (df["Salary"] >= 60000))
]
```
