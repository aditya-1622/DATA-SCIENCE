```python
df = pd.DataFrame({
    "order_id": [101,102,103,104,105,106,107],
    "customer": ["A","B","C","A","B","C","D"],
    "city": ["Bangalore","Delhi","Mumbai","Bangalore","Delhi","Mumbai","Chennai"],
    "amount": [500,200,300,450,150,350,400],
    "category": ["Electronics","Clothing","Electronics","Clothing","Electronics","Clothing","Electronics"],
    "order_date": [
        "2024-01-10",
        "2024-02-15",
        "2024-03-20",
        "2024-01-25",
        "2024-02-05",
        "2024-03-15",
        "2024-01-05"
    ]
})

df
```

```python
# Q1. 📅 Date Transformation and Feature Extraction

# The order_date column is currently stored as a string.

# Your task is to:

# Convert the order_date column into a proper datetime format.
# From this datetime column, create the following new columns:
# year → year of the order
# month → numeric month (1–12)
# day → day of the month
# month_name → full month name (e.g., January)
# is_weekend → True if the order was placed on Saturday or Sunday, else False

#convert to date time format (very imp before using .dt)
df["order_date"] = pd.to_datetime(df["order_date"])

#Extract component.
df["year"] = df["order_date"].dt.year
df["month"]= df["order_date"].dt.month
df["day"] = df["order_date"].dt.day

#Extract readable month name jan feb
df["month_name"]= df["order_date"].dt.month_name()

#check if weeekend
df["is_weekend"] = df["order_date"].dt.dayofweek >= 5
df
```

```python
# Q2. 🔧 Applying Business Logic Using apply()

# Create a new column called spend_type based on the amount spent by each customer.

# Rules:

# If amount is greater than or equal to 400, label as "High"
# If amount is between 200 and 399, label as "Medium"
# If amount is less than 200, label as "Low"

# Use the apply() function with a lambda expression.

#define a function for bussiness logic.
def categorize_spend(x):
  if x >= 400:
    return "high"
  elif x >= 200:
      return "medium"
  else:
    return "low"

    #apply the function to the column.
  df["spend_type"] = df["amount"].apply(categorize_spend)
  df
```

```python
# Q3. 🔄 Value Transformation using map() and replace()

# You need to standardize and encode categorical values.

# Steps:

# Create a new column city_code by mapping:
# Bangalore → BLR
# Delhi → DEL
# Mumbai → MUM
# Chennai → CHE
# Modify the existing category column by replacing:
# Electronics → E
# Clothing → C

#map city name s--> codes 1 to 1 mapping
city_map = {
     "Bangalore": "BLR",
     "Delhi": "DEL",
     "Mumbai": "MUM",
     "Chennai": "CHE"

}
df["city_code"] = df ["city"].map(city_map)

#replace values inside cols (cleaning/encoding)
df["category"] = df ["category"].replace({
    "Electronics" : "E",

    'Clothing': "C"

    })
df
```

```python
# Q4. Reshaping Data using Pivot

# You are asked to analyze total spending by each customer across categories.

# Create a pivot table such that:

# Rows represent customer
# Columns represent category
# Values represent the total amount spent

pivot_df = df.pivot_table(
         index= "customer",  #rows
         columns = "category", #cols
         values="amount",  #values to aggregation
         aggfunc="sum"    #agg function
)
df
```

```python
# Q5. Converting Wide Data Back to Long Format (Melt)

# Using the pivot table created in Q4:

# Convert the wide-format pivot table back into a long-format dataset
# The final output should contain:
# customer
# category
# amount

pivot_df = df.pivot_table(
    index="customer",        # rows
    columns="category",      # columns
    values="amount",         # values to aggregate
    aggfunc="sum"            # aggregation function
)
df
```

```python
# Q6. Efficient Computation using Vectorization

# Create a new column called discounted_amount.

# Rules:

# If amount is greater than 300, apply a 10% discount
# Otherwise, keep the amount unchanged

# Important:
# Do NOT use apply() — use vectorized operations.

 #First apply 10% discount to all
df["discounted_amount"] = df["amount"] * 0.9

# Then overwrite where condition is not met
df.loc[df["amount"] <= 300, "discounted_amount"] = df["amount"]
df
```

```python
# Q7. End-to-End Data Transformation Pipeline

# Perform the following sequence of operations:

# Convert order_date to datetime (if not already done)
# Extract the month name into a new column
# Create a pivot table where:
# Rows = month
# Columns = category
# Values = total amount
# Convert this pivot table back into long format using melt()

# Final output should look like:

# month | category | total_amount

# Extract month name
df["month"] = df["order_date"].dt.month_name()

# Pivot: month vs category → total amount
temp = df.pivot_table(
    index="month",
    columns="category",
    values="amount",
    aggfunc="sum"
)

# Melt back to long format
q7 = pd.melt(
    temp.reset_index(),
    id_vars=["month"],
    var_name="category",
    value_name="total_amount"
)
df
```

```python
# Q8. A business analyst wants to understand revenue trends across cities and months.

# Your task:

# Extract the month from order_date
# Calculate total revenue for each city per month
# Sort the result in descending order of revenue

# Final output should help identify:

# Which city generates the highest revenue
# In which month

# Group by city and month → total revenue
q8 = df.groupby(["city", "month"])["amount"].sum().reset_index()

# Sort highest revenue first
q8 = q8.sort_values(by="amount", ascending=False)
df
```

```python
# Q9. Pivot with Multiple Aggregations (Advanced Pivot)

# You previously created a pivot table for total spend. Now, the business wants more insights per customer.

# Create a pivot table where:

# Rows represent customer
# Columns represent category
# Values represent amount

# But this time:

# Calculate both:
# Total amount spent (sum)
# Average amount spent (mean)

# Your output should show multiple aggregated values for each category.


q9 = df.pivot_table(
    index="customer",
    columns="category",
    values="amount",
    aggfunc=["sum", "mean"]  # multiple aggregations
)
df
```

```python
# Q10. Filtering + Transformation (Real-world logic)

# A marketing team only wants to analyze high-value transactions.

# Steps:

# Filter the dataset to include only rows where amount ≥ 300
# From this filtered data:
# Create a new column high_value_flag = "Yes"
# Return only:
# customer, city, amount, high_value_flag


# Filter rows
q10 = df[df["amount"] >= 300].copy()

# Add flag column
q10["high_value_flag"] = "Yes"

# Select required columns
q10 = q10[["customer", "city", "amount", "high_value_flag"]]

df
```

```python
# Q11. Conditional Transformation without apply()

# Create a new column order_size based on amount:

# Rules:

# amount ≥ 400 → "Large"
# 200 ≤ amount < 400 → "Medium"
# < 200 → "Small"

# Constraint:

# Do NOT use apply()
# Use vectorized logic (np.select or boolean masking)

conditions = [
    df["amount"] >= 400,
    (df["amount"] >= 200) & (df["amount"] < 400),
    df["amount"] < 200
]

choices = ["Large", "Medium", "Small"]

# Apply vectorized conditional logic, providing a string default value
df["order_size"] = np.select(conditions, choices, default=None)
df
```

```python
# Q12. Time-based Aggregation (Important Real Use Case)

# The business wants to track monthly revenue trends.

# Steps:

# Extract month_name from order_date
# Group the data by month_name
# Calculate total amount per month
# Sort months in calendar order (NOT alphabetical)

# Final output should clearly show:

# month_name | total_revenue


# Group by month
q12 = df.groupby("month_name")["amount"].sum().reset_index()

# Define correct calendar order (important!)
month_order = ["January", "February", "March", "April", "May", "June",
               "July", "August", "September", "October", "November", "December"]

# Convert to ordered categorical for sorting
q12["month_name"] = pd.Categorical(q12["month_name"], categories=month_order, ordered=True)

# Sort properly (not alphabetical)
q12 = q12.sort_values("month_name")
df
```
