```python
#importing important libraries..
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the data
df = pd.read_csv ("marketing_and_purchases.csv")
df.head()
```

```python
#basic probabality analysis.

n = len(df)
p_open = df['email_opened'].mean()
p_click = df['email_opened'].mean()
print(f'p (open) = {p_open:.3f}')
print(f'(click) = {p_click:.3f}')
```

```python
#multiplicatioon and complement rules
p = p_open
print(f'P(both open) = {p*p : .3f}')
print(f'P(neither opens) = {(1-p)*(1-p) : .3f}')
print(f'P(at least one opens) = {(1-(1-p)*(1-p)) : .3f}')
```

```python
# Conditional Probability
p_pastry_all = df['bought_pastry'].mean()
p_pastry_coffee = df[df['bought_coffee'] == 1]['bought_pastry'].mean()

print(f"P(pastry) = {p_pastry_all:.3f}")
print(f"P(pastry | coffee) = {p_pastry_coffee:.3f}")
print(f"Lift = {p_pastry_coffee / p_pastry_all:.3f}x")
```

```python
# bayes theroem (two machine example in the document)
f = pd.read_csv("factory_quality.csv")
df.head()
```

```python
# grouped data
f.groupby('machine')['defective'].agg(['count', 'mean'])
```

```python
# Bayes Theorem Example by hand
P_A, P_B = 0.60, 0.40
P_def_A, P_def_B = 0.02, 0.05

P_def = P_A * P_def_A + P_B * P_def_B
P_B_def = (P_B*P_def_B) / P_def
print(f'P(defective) = {P_def:.3f}')
print(f'P(B | defective) = {P_B_def:.3f}')
```

```python
# Verify empirically from the data
defectives = f[f['defective'] == 1]
empirical = (defectives['machine'] == 'B').mean()
print(f"P(B | defective) = {empirical:.3}")
```
