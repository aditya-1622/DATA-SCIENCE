```python
# Part 2 :
# CRUD operations : Create, Read, Update, Delete
```

```python
import mysql.connector
conn = mysql.connector.connect(
    host = "localhost", # if using google colab : server ip/domain (host="127.0.0.1")
    user = "root",
    password = "Makkar123!!", # workbench password # enter your own password!
    database = "company_db"
)
```

```python
cursor = conn.cursor()
```

```python
cursor.execute("""
INSERT INTO employee
(Dept, ID, Salary)
VALUES
('Admin', 17, 1000)
""")
```

```python
# Is this row already inserted in the table or not?
conn.commit() # save button!
```

```python
cursor.execute(
    "SELECT * FROM employee"
)

rows = cursor.fetchall()
```

```python
rows
```

```python
# Update, Delete command on the same table and verify!
```

```python
cursor.execute("""
UPDATE employee
SET salary=90000
WHERE dept='Admin'
""")

conn.commit()
```

```python
cursor.execute(
    "SELECT * FROM employee"
)

rows = cursor.fetchall()
```

```python
rows
```

```python
cursor.execute("""
DELETE FROM employee
WHERE dept='Admin'
""")

conn.commit()
```

```python
cursor.execute(
    "SELECT * FROM employee"
)

rows = cursor.fetchall()
```

```python
rows
```

```python
cursor.close()
conn.close()
```
