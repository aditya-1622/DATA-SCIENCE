```python
# Importing important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the data
df = pd.read_csv("C:/Users/jaipu/Documents/Masai/New/Probability_&_Statistics/Content/Data/marketing_and_purchases.csv")
df.head()
```

```python
# Basic Probability Analysis
n = len(df)
p_open = df['email_opened'].mean()
p_click = df['email_clicked'].mean()
print(f'P(open) = {p_open:.3f}')
print(f'P(click) = {p_click:.3f}')
```

```python
# Multiplication & Complement Rules
p = p_open
print(f'P(both open) = {p*p : .3f}')
print(f'P(neither opens) = {(1-p)*(1-p) : .3f}')
print(f'P(at least one opens) = {(1-(1-p)*(1-p)) : .3f}')
```

```python
# Conditional Probability
p_pastry_all = df['bought_pastry'].mean()
p_pastry_coffee = df[df['bought_coffee'] == 1]['bought_pastry'].mean()

print(f"P(pastry) = {p_pastry_all:.3f}")
print(f"P(pastry | coffee) = {p_pastry_coffee:.3f}")
print(f"Lift = {p_pastry_coffee / p_pastry_all:.3f}x")
```

```python
# Bayes Theorem (Two Machines Example in the document)
df = pd.read_csv("C:/Users/jaipu/Documents/Masai/New/Probability_&_Statistics/Content/Data/factory_quality.csv")
df.head()
```

```python
# Grouped Data
df.groupby('machine')['defective'].agg(['count', 'mean'])
```

```python
# Bayes Theorem Example by hand
P_A, P_B = 0.60, 0.40
P_def_A, P_def_B = 0.02, 0.05

P_def = P_A * P_def_A + P_B * P_def_B
P_B_def = (P_B*P_def_B) / P_def
print(f'P(defective) = {P_def:.3f}')
print(f'P(B | defective) = {P_B_def:.3f}')
```

```python
# Verify empirically from the data
defectives = df[df['defective'] == 1]
empirical = (defectives['machine'] == 'B').mean()
print(f"P(B | defective) = {empirical:.3}")
```
